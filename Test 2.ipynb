```python
#Question 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#Define dy/dt=-2yt
def eqn(y,t):
    return (-2*y*t)

#Initial condition
y0 = 1

t = np.linspace(0,2,100) #time points where the solution is being evaluated

solution = odeint(eqn, y0, t) #solve ODE by odeint

#Plot solution graphically
plt.figure(figsize=(10,8))
plt.plot(t, solution, label='y(t)')
plt.title('Solution of dy/dt=-2yt given y(0)=1')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
